##### 7. str.index(sub[, start[, end]]) and str.rindex() 


* index : Returns the lowest index (first occurrence) of the substring sub in the string.

* rindex : returns the highest index of the substring inside the string if the substring is found. 
Otherwise, it raises ValueError.

```python
# Using Index
text = "hello world hello"
print(text.index("hello"))   
print(text.index("hello", 1))
```

```python
# Using rindex
print('Python'.rindex('th'))  
```

```python
text = "python programming"

try:
    pos = text.rindex("java")
    print("Found at:", pos)
except ValueError:
    print("Substring not found")
```

##### 15.str.center(width, fillchar=' '), str.ljust(), str.rjust()

* str.center(width, fillchar=' ')

```python
text = "Hello"

# Center with spaces
print(text.center(11))         

# Center with custom fill character
print(text.center(11, '*'))   

# Width smaller than string length
print(text.center(3, '-'))    
```

* str.ljust()

```python
text = "Hello"

# Left align with spaces
print(text.ljust(10))         

# Left align with custom fill character
print(text.ljust(10, '-'))     

# Width smaller than string length
print(text.ljust(3, '.'))      
```

* str.rjust(width, fillchar=' ')

```python
text = "Hello"

# Right align with spaces
print(text.rjust(10))          

# Right align with custom fill character
print(text.rjust(10, '.'))    

# Width smaller than string length
print(text.rjust(3, '*'))     
```

# Input Method for String

* Basic String Input

```python
name = input("Enter your name: ")
print("Hello,", name)
```

* Numeric Input (conversion needed)

```python
age = input("Enter your age: ")
print(type(age))   

# Convert to integer
age = int(age)
print("Next year you will be", age + 1)
```

* Multiple Inputs

```python
data = input("Enter two words: ")  
word1, word2 = data.split()
print("First:", word1)
print("Second:", word2)
```

* Custom Separator

```python
data = input("Enter values separated by commas: ")  
items = data.split(",")
print(items)   
```

## Input Method for Integer

* Basic Integer Input

```python
num = int(input("Enter an integer: "))
print("You entered:", num)
print("Type:", type(num))
```

* Arithmetic with Integer Input

```python
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
print("Sum =", x + y)
```

* Handling Multiple Integers

```python
nums = input("Enter numbers separated by space: ")   
nums_list = list(map(int, nums.split()))
print(nums_list)   
```

* Error Handling (Invalid Input)

```python
try:
    num = int(input("Enter an integer: "))
    print("Square:", num ** 2)
except ValueError:
    print("Oops! That wasn’t a valid integer.")
```
